### Проверим, что эффективность обогащения рассчитана правильно

Вычислим эффективность обогащения для `rougher.output.recovery`

Напишем формулу вычисления эффективности обогащения

```python
def recovery(c, f, t):
    recovery = (c * (f - t)) / (f * (c - t)) * 100
    return recovery
```

Сохраним в переменных необходимы для вычислений признаки

```python
c = df_train['rougher.output.concentrate_au']
f = df_train['rougher.input.feed_au']
t = df_train['rougher.output.tail_au']
```

Вычислим эффективность обогащения

```python
recovery = recovery(c, f, t)
```

Найдем `MAE` между нашими расчетами и значением признака

```python
mean_absolute_error(df_train['rougher.output.recovery'], recovery).round(3)
```

Значение `MAE` - 0 говорит о том, что средняя разница между расчетным и фактическим значением эффективности обогащения отсутствует. Эффективность обогащения рассчитана верно

#### Мультиколлинеарность

Для решения проблемы мультиколлинеарности используем алгоритм Брона-Кербоша для поиска максимальных клик в графе, где каждый признак является вершиной, а рёбра соединяют только те вершины, у которых коэффициент корреляции меньше заданного порога. Для этого будем использовать библиотеку `NetworkX`:
 1. Вычислим модуль корреляции Пирсона между каждым признаком и целевым признаком и сохраним в датафрейм
   + Так как у нас два целевых признака, связанных друг с другом, но для итоговой метрики, `sMAPE` финального концентрата важнее в три раза, мы будем ориентироваться на корреляцию признаков именно с эффективностью обогащения финального концентрата
 2. Создадим матрицу модуля корреляции для всех признаков
 3. Выберем границу корреляции и сохраним в виде ребер только те признаки, корреляция между которыми ниже заданной границы
 4. Создадим граф и добавим в него ребра
 5. Произведем поиск максимальных клик, и среди клик наибольшего размера найдем оптимальную с точки зрения суммарной корреляции с целевым признаком
 6. Решим проблему мультиколлинеарности, оставим в выборках только признаки из оптимальной клики

Рассчитаем коэффициент корреляции пирсона между признаками и целевым признаком

```python
corr_target = pd.DataFrame(columns=['final_recovery_corr'])
```

```python
for col in cols:
    corr = abs(features_train[col].corr(target_train['final.output.recovery']))
    corr_target.loc[col] = corr
```

Создадим матрицу корреляций для всех признаков в выборке

```python
corr_matrix = features_train.corr().abs()
```

Установим границу максимально допустимой корреляции

```python
threshold=0.7
```

Напишем функцию для поиска оптимальной клики

```python
def find_best_clique(corr_target=corr_target, corr_matrix=corr_matrix, threshold=0.7, calculate=True):
    
    #Создадим пустой список для добавления ребер признаков
    edges = []
    
    #Отберем пары признаков с корреляцией меньше порога
    for column in corr_matrix.columns:
        for index in corr_matrix.index:
            if corr_matrix.loc[index, column] < threshold:
                edges.append((index, column))
                
    #Создадим граф
    G = nx.Graph()
    
    #Добавим в него ребра
    for edge in edges:
        G.add_edge(*edge)
     
    #Найдем максимальные клики
    cliques = list(nx.find_cliques(G))
    if calculate == True:
        print(f'{len(cliques)} клик найдено')
    
    #Рассчитаем длину каждой клики
    lens = [len(clique) for clique in cliques]
    
    #Выведем сколько всего клик разного размера удалось найти
    if calculate == True:
        print(pd.Series(lens).value_counts())
    
    #Определим размер наибольших клик
    max_size = max(lens)
    
    #Отберем клики максимального размера
    max_cliques = [clique for clique in cliques if len(clique) == max_size]
    
    #Подсчитаем для каждой клики максимального размера сумму модулей коэффициентов корреляции
    clique_corr_sums = [corr_target.loc[clique, 'final_recovery_corr'].sum() for clique in max_cliques]
    
    #Найдем оптимальную клику с точки зрения суммарной корреляции с целевым признаком
    best_clique = cliques[np.argmax(clique_corr_sums)]
      
    return best_clique
```

### Напишем функции для вычисления sMAPE 

Так как если при сумме модулей получается 0, то и модуль разности будет равен 0, а значит корректно будет заменить 0 в знаменателе на 1, для избегания ошибки связанной с делением на 0

```python
uniordef sMAPE(t, p):
    """
    Вычисляет симметричную среднюю абсолютную процентную ошибку (SMAPE).

    Параметры:
    t (pd.Series): Набор целевых значений.
    p (pd.Series): Набор прогнозируемых значений.

    Возвращает:
    float: Значение SMAPE.
    """
    t = np.array(t)
    p = np.array(p)
    #Отдельно посчитаем знаменатель
    divisor = np.abs(t) + np.abs(p)
    #И заменим получившиеся 0 на 1 для избегания ошибок связанных с делением на 0
    divisor[divisor == 0] = 1
    smape = 1 / len(t) * np.sum((np.abs(t - p) / (divisor / 2) * 100))
    return smape
```

Напишем функцию для вычисления итогового sMAPE

```python
def final_smape(smape_rougher, smape_second):
    """
    Вычисляет итоговую метрику SMAPE на основе SMAPE для этапов "rougher" и "second".

    Параметры:
    smape_rougher (float): Значение SMAPE для этапа "rougher".
    smape_second (float): Значение SMAPE для этапа "second".

    Возвращает:
    float: Итоговое значение SMAPE, взвешенное на основе значений для разных этапов.
    """
    
    return 0.25 * smape_rougher + 0.75 * smape_second
```

### Проверка на тестовой выборке

Напишем функцию проводящую финальное тестирование на заданной выборке

```python
def final_test(features_train, features_test, rougher_params={'random_state': 1},
               final_params={'random_state': 1}, *, model=ElasticNet):
    """
    Выполняет окончательное тестирование модели на тестовых данных.

    Аргументы:
        features_train (numpy.ndarray): Данные признаков для обучения модели.
        features_test (numpy.ndarray): Данные признаков для тестирования модели.

    Возвращает:
        None: Выводит на экран рассчитанное окончательное значение sMAPE.

    """

    model_for_rougher = model(**rougher_params)
    model_for_final = model(**final_params)
    
    final_pipe.steps[1] = ('model', model_for_rougher)
    final_pipe.fit(features_train, target_train['rougher.output.recovery'])
    predicted_rougher = final_pipe.predict(features_test)
    
    final_pipe.steps[1] = ('model', model_for_final)
    final_pipe.fit(features_train, target_train['final.output.recovery'])
    predicted_final = final_pipe.predict(features_test)


        
    #Высчитаем отдельно sMAPE для двух целевых признаков    
    smape_1 = sMAPE(target_test['rougher.output.recovery'], predicted_rougher)
    smape_2 = sMAPE(target_test['final.output.recovery'], predicted_final)
    
    #Высчитаем финальное sMAPE
    total = final_smape(smape_1, smape_2)
    
    print(f'Итоговое sMAPE: {total.round(3)}%')
```

Протестируем модель `Elastic Net` на тестовой выборке

```python
final_test(features_train_3, features_test_3,
           best_params_elastic_rougher, best_params_elastic_final, model=ElasticNet)
```
