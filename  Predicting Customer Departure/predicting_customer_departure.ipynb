#### Функции для вывода информации

```python
def gaps(data):
    '''
    Функция выводит количество пропусков и их долю в каждом столбце
    '''
    for col in data.columns:
            gaps = data[col].isna().sum()
            if gaps > 0:
                print(f'Количество пропусков в столбце\
                      {col} ({data[col].dtypes}): {gaps},\
                      {(gaps / len(data)):.2%} от всех данных')
```

```python
def get_dict(f_train, t_train):
    '''
    Функция возвращает словарь с образцами данных для обучения и тестирования модели.
    '''
    samples = {
        'features_train': f_train,
        'features_test': f_test,
        'features_valid': f_valid,
        'target_train': t_train,
        'target_test': t_test,
        'target_valid': t_valid,    
    }
    return samples
```

```python
def get_size(samples):
    '''
    Функция выводит размеры выборок и соотношение классов (если выборка является pd.Series).
    '''
    for name, sample in samples.items():
        print(f'- Размер выборки {name}: {sample.shape}')
        if isinstance(sample, pd.Series):
            class_counts = sample.value_counts()
            prop = (class_counts / class_counts[1]).round(1)
            print(f'Соотношение классов в выборке: {prop[0]}:{prop[1]}')
```

#### Создадим пайплайн предобработки данных

Предобработка числовых переменных

```python
numeric = ['credit_score', 'age', 'balance', 'estimated_salary', 'tenure', 'num_of_products']
numeric_transformer = StandardScaler()
```

Объединим в пайплайн заполнение пропусков и масштабирование

```python
simple_imputer = SimpleImputer(strategy='median')
```

```python
pipe_num = Pipeline(steps=[('misses', simple_imputer),
                           ('scaler', numeric_transformer)
                          ])
```

Предобработка категориальных переменных

```python
categorical = ['geography', 'gender', 'has_cr_card', 'is_active_member']
categoriсal_transformer = OneHotEncoder(drop='first')
```

Создадим комбинированный препроцессор для числовых и категориальных переменных

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categoriсal_transformer, categorical),
        ('num', pipe_num, numeric)
    ])
```

Проверим работу препроцессора

```python
f_train_processed = preprocessor.fit_transform(f_train)
```

```python
pd.DataFrame(f_train_processed, columns=feature_names).info()
```

Сформируем финальный пайплайн оставив место для вставки метода ресэмплирования и модели

```python
final_pipe = Pipeline(steps=[('preproc', preprocessor),
                             ('resample', None),
                             ('model', None)])
```

## Исследование задачи

### Подберем гиперпараметры для моделей классификации

Определимся с моделью

```python
models = {
    'Logistic Regression': LogisticRegression(random_state=1),
    'Random Forest': RandomForestClassifier(random_state=1),
    'Decision Tree': DecisionTreeClassifier(random_state=1)
}
```

Напишем функцию для получения столбца с основными метриками

```python
def get_metrics(predictions, probabilities_one_valid,*, t_valid=t_valid):
    f1 = f1_score(t_valid, predictions).round(3)
    roc_auc = roc_auc_score(t_valid, probabilities_one_valid).round(3)
    accuracy = accuracy_score(t_valid, predictions).round(3)
    recall = recall_score(t_valid, predictions).round(3)
    precision = precision_score(t_valid, predictions).round(3)
    new_col = f1, roc_auc, accuracy, recall, precision
    return new_col
```

Напишем функцию, которой будем передавать метод ресэмплирования/значение параметра взвешивания классов и лучшие параметры, а обратно получать столбец с метриками

```python
def metrics_in_table(best_params, resample=None, *, class_weight=None, f_valid=f_valid, t_valid=t_valid):
    '''
    Функция вычисляет метрики и возвращает их в виде столбца.
    '''
    model = RandomForestClassifier(random_state=1, **best_params, class_weight=class_weight)
    final_pipe.steps[1] = ('resample', resample)
    final_pipe.steps[2] = ('random_forest', model)
    
    final_pipe.fit(f_train, t_train)
    predictions = final_pipe.predict(f_valid)
    predict_proba = final_pipe.predict_proba(f_valid)
    probabilities_one_valid = predict_proba[:, 1]
    
    return get_metrics(predictions, probabilities_one_valid)
```

```python
score_table['defolt'] = metrics_in_table(best_params_defolt)
```

### Подбор порога

Напишем функцию подбора оптимального порога

```python
def find_best_threshold(best_params,*, class_weight=None, resample=None, f_valid=f_valid, t_valid=t_valid):
    '''
    Функция подбирает оптимальный порог классификации и вычисляет метрики.
    '''
    
    model = RandomForestClassifier(random_state=1, **best_params, class_weight=class_weight)
    final_pipe.steps[1] = ('resample', resample)
    final_pipe.steps[2] = ('random_forest', model)
    final_pipe.fit(f_train, t_train)
    predictions = final_pipe.predict(f_valid)
    predict_proba = final_pipe.predict_proba(f_valid)
    probabilities_one_valid = predict_proba[:, 1]
    
    best_f1 = 0
    best_threshold = 0
    for threshold in np.arange(0, 0.9, 0.02):
        predicted_valid = probabilities_one_valid > threshold
        f1 = f1_score(t_valid, predicted_valid).round(3)
        if f1 > best_f1:
            best_f1 = f1
            best_threshold = threshold
    predicted_valid = probabilities_one_valid > best_threshold

    print(f'Лучшее f1_score={best_f1} при пороге={best_threshold}')
    return get_metrics(predicted_valid, predicted_valid, t_valid=t_valid)
```
